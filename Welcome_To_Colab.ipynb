```
from presidio_analyzer import AnalyzerEngine, PatternRecognizer, Pattern, RecognizerResult
from presidio_anonymizer import AnonymizerEngine
from presidio_analyzer.nlp_engine import NlpEngineProvider
from flair.data import Sentence
from flair.nn import Classifier
import pandas as pd
import openpyxl
import json
```

```

# Load NER tagger
ner_tagger = Classifier.load("flair/ner-english-large")
```

```

# Configure NLP engine
nlp_config = {
    "nlp_engine_name": "spacy",
    "models": [{"lang_code": "en", "model_name": "en_core_web_lg"}]
}

# Initialize NLP engine
provider = NlpEngineProvider(nlp_configuration=nlp_config)
nlp_engine = provider.create_engine()

# Initialize analyzer and anonymizer
analyzer = AnalyzerEngine(nlp_engine=nlp_engine)
anonymizer = AnonymizerEngine()

# Define column-specific anonymization rules
column_anonymization_rules = {
    "Name": "<PERSON>",
    "Company": "<ORGANIZATION>",
    "Email": "<EMAIL_ADDRESS>",
    "Phone": "<PHONE_NUMBER>",
    "Address": "<STREET_ADDRESS>",
}

# Dictionary to store mappings for de-anonymization
anonymization_mappings = {col: {} for col in column_anonymization_rules.keys()}
free_text_mappings = {}

def anonymize_column_based_on_name(column_name, value):
    """
    Anonymize a value based on the column name and store the mapping.
    """
    if column_name in column_anonymization_rules:
        anonymized_value = column_anonymization_rules[column_name]
        anonymization_mappings[column_name][value] = anonymized_value
        return anonymized_value
    return value

def detect_locations_with_flair(text):
    """
    Detect entities in text using Flair.
    """
    sentence = Sentence(text)
    ner_tagger.predict(sentence)

    location_entities = []
    for entity in sentence.get_spans("ner"):
        if entity.tag in ["LOC", "ORG", 'PER']:
            entity_type = "LOCATION" if entity.tag == "LOC" else "ORGANIZATION" if entity.tag == "ORG" else "PERSON"
            location_entities.append(RecognizerResult(
                entity_type=entity_type,
                start=entity.start_position,
                end=entity.end_position,
                score=entity.score
            ))
    return location_entities

def analyze_and_anonymize(text):
    """
    Anonymize text using Presidio and Flair, and store mappings.
    """
    presidio_results = analyzer.analyze(
        text=text,
        entities=["PERSON", "EMAIL_ADDRESS", "PHONE_NUMBER", "STREET_ADDRESS"],
        language="en",
    )

    flair_results = detect_locations_with_flair(text)
    all_results = flair_results + presidio_results

    anonymized_text = anonymizer.anonymize(text, all_results).text

    # Store mapping for free-text fields
    if anonymized_text != text:
        free_text_mappings[anonymized_text] = text

    return anonymized_text

def anonymize_excel(input_file, output_file):
    """
    Anonymize an Excel file using a hybrid approach.
    """
    # Load Excel file
    wb = openpyxl.load_workbook(input_file)

    for sheet_name in wb.sheetnames:
        sheet = wb[sheet_name]

        # Get header row (column names)
        headers = [cell.value for cell in sheet[1]]

        # Iterate through rows (starting from the second row)
        for row in sheet.iter_rows(min_row=2):
            for idx, cell in enumerate(row):
                column_name = headers[idx]  # Get the column name
                if cell.value and isinstance(cell.value, str):
                    if column_name in column_anonymization_rules:
                        # Column-based anonymization
                        cell.value = anonymize_column_based_on_name(column_name, cell.value)
                    else:
                        # Entity-based anonymization for free-text columns
                        cell.value = analyze_and_anonymize(cell.value)

    # Save anonymized data
    wb.save(output_file)
    print(f"Anonymized file saved as {output_file}")

    # Save mappings to JSON files
    with open("column_mappings.json", "w") as f:
        json.dump(anonymization_mappings, f)
    with open("free_text_mappings.json", "w") as f:
        json.dump(free_text_mappings, f)
    print("Mappings saved to JSON files.")

def de_anonymize_excel(input_file, output_file):
    """
    De-anonymize an Excel file using stored mappings.
    """
    # Load mappings from JSON files
    with open("column_mappings.json", "r") as f:
        column_mappings = json.load(f)
    with open("free_text_mappings.json", "r") as f:
        free_text_mappings = json.load(f)

    # Load anonymized Excel file
    wb = openpyxl.load_workbook(input_file)

    for sheet_name in wb.sheetnames:
        sheet = wb[sheet_name]

        # Get header row (column names)
        headers = [cell.value for cell in sheet[1]]

        # Iterate through rows (starting from the second row)
        for row in sheet.iter_rows(min_row=2):
            for idx, cell in enumerate(row):
                column_name = headers[idx]  # Get the column name
                if cell.value and isinstance(cell.value, str):
                    if column_name in column_mappings:
                        # De-anonymize column-based data
                        for original_value, anonymized_value in column_mappings[column_name].items():
                            if cell.value == anonymized_value:
                                cell.value = original_value
                                break
                    else:
                        # De-anonymize free-text data
                        if cell.value in free_text_mappings:
                            cell.value = free_text_mappings[cell.value]

    # Save de-anonymized data
    wb.save(output_file)
    print(f"De-anonymized file saved as {output_file}")
```

```
anonymize_excel("sensitive.xlsx", "anonymized.xlsx")

# Run De-Anonymization
de_anonymize_excel("anonymized.xlsx", "de_anonymized.xlsx")
```
